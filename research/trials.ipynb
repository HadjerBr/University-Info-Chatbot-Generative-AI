```python
from langchain_community.document_loaders import PyPDFLoader, DirectoryLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
# Extract Data From the PDF Files


def load_pdf_file(data):
    loader = DirectoryLoader(
        data,
        glob="*.pdf",
        loader_cls=PyPDFLoader
    )
    
    documents = loader.load()
    return documents
```

```python
%pwd
```

```python
extracted_data = load_pdf_file(data='Data/')
```

```python
# extracted_data
```

```python
import re
import unicodedata

def clean_text(text: str) -> str:
    # Normalize weird unicode 
    text = unicodedata.normalize("NFKC", text)

    # Remove trailing spaces before newlines: 
    text = re.sub(r"[ \t]+\n", "\n", text)

    # Collapse many blank lines into a single blank line
    text = re.sub(r"\n{3,}", "\n\n", text)

    # Collapse multiple spaces inside a line
    text = re.sub(r"[ \t]{2,}", " ", text)

    # Strip leading/trailing whitespace
    text = text.strip()

    return text
```

```python
cleaned_docs = []

for doc in extracted_data:
    doc.page_content = clean_text(doc.page_content)
    cleaned_docs.append(doc)
```

```python
cleaned_docs
```

```python
# Chunking
def text_split(cleaned_docs):
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=500,
        chunk_overlap=50
    )
    
    text_chunks = text_splitter.split_documents(cleaned_docs)
    return text_chunks
```

```python
text_chunks = text_split(cleaned_docs)
print("Length of Text Chunks:", len(text_chunks))
```

```python
text_chunks
```

```python
from langchain_huggingface.embeddings import HuggingFaceEmbeddings
```

```python
# Download the Embeddings from Hugging Face
def download_hugging_face_embeddings():
    embeddings = HuggingFaceEmbeddings(
        model_name="sentence-transformers/all-MiniLM-L6-v2" # 384 dimensional dense vector space
    )
    return embeddings
```

```python
embeddings = download_hugging_face_embeddings()
```

```python
from pinecone.grpc import PineconeGRPC as Pinecone
from pinecone import ServerlessSpec


pc = Pinecone(PINECONE_API_KEY)

index_name = "campusmate-db"

if not pc.has_index(index_name):
    pc.create_index(
        name=index_name,
        vector_type="dense",
        dimension=384,
        metric="cosine",
        spec=ServerlessSpec(
            cloud="aws",
            region="us-east-1"
        ),
        deletion_protection="disabled",
        tags={
            "environment": "development"
        }
    )
```

```python
from langchain_pinecone import PineconeVectorStore

# Create embeddings for each chunk and upload them to the Pinecone index
docsearch = PineconeVectorStore.from_documents(
    documents=text_chunks,
    index_name=index_name,
    embedding=embeddings,
)
```

```python
# Load an existing Pinecone index

from langchain_pinecone import PineconeVectorStore

docsearch = PineconeVectorStore.from_existing_index(
    index_name=index_name,
    embedding=embeddings
)
```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    model="gpt-4o-mini",   
    temperature=0.6,
    max_tokens=500
)
```

```python
import langchain_classic
from langchain_classic.chains import create_retrieval_chain
from langchain_classic.chains.combine_documents import create_stuff_documents_chain
from langchain_core.prompts import ChatPromptTemplate
```

```python

# LangChain Retrieval + Prompt Imports


system_prompt = (
    "You are an assistant that answers questions based only on the provided context.\n"
    "Do NOT use prior knowledge or make up answers.\n"
    "Use the retrieved information to answer the question.\n"
    "If the answer is not in the context, respond with: 'I don't know.'\n\n"
    "{context}"
)

# Build a chat prompt template
prompt = ChatPromptTemplate.from_messages([
    ("system", system_prompt),
    ("human", "{input}")
])

```

```python
qa_chain = create_stuff_documents_chain(llm, prompt)
rag_chain = create_retrieval_chain(retriever, qa_chain)
```

```python
response = rag_chain.invoke({"input":"Is it mendatory to take an English exam to get to Harvard College for international students?"})
print(response["answer"])
```
